```python
"""
Function
--------
simulate_prizedoor

Generate a random array of 0s, 1s, and 2s, representing
hiding a prize between door 0, door 1, and door 2

Parameters
----------
nsim : int
    The number of simulations to run

Returns
-------
sims : array
    Random array of 0s, 1s, and 2s

Example
-------
>>> print simulate_prizedoor(3)
array([0, 0, 2])
"""
def simulate_prizedoor(nsim):
    #compute here
    answer = np.random.randint(3, size=nsim)
    return answer
#your code here
```

```python
"""
Function
--------
simulate_guess

Return any strategy for guessing which door a prize is behind. This
could be a random strategy, one that always guesses 2, whatever.

Parameters
----------
nsim : int
    The number of simulations to generate guesses for

Returns
-------
guesses : array
    An array of guesses. Each guess is a 0, 1, or 2

Example
-------
>>> print simulate_guess(5)
array([0, 0, 0, 0, 0])
"""
#your code here

def simulate_guess(nsim):
    return np.random.randint(3, size=nsim)

simulate_guess(20)
```

```python
"""
Function
--------
goat_door

Simulate the opening of a "goat door" that doesn't contain the prize,
and is different from the contestants guess

Parameters
----------
prizedoors : array
    The door that the prize is behind in each simulation
guesses : array
    THe door that the contestant guessed in each simulation

Returns
-------
goats : array
    The goat door that is opened for each simulation. Each item is 0, 1, or 2, and is different
    from both prizedoors and guesses

Examples
--------
>>> print goat_door(np.array([0, 1, 2]), np.array([1, 1, 1]))
>>> array([2, 2, 0])
"""
#your code here

def goat_door(prizedoors, guesses):
    
    #generate same size array as prizedoors
    goat = np.random.randint(3, size=prizedoors.size)
    
    while True:
        #compares goat to prizedoors and guess and return boolean array
        bad = (goat == prizedoors) | (goat == guesses)
        #if all False return goat
        if not bad.any():
            return goat
        goat[bad] = np.random.randint(3, size=bad.sum())
        
```

```python
print(goat_door(np.array([0,0,1]), np.array([1,2,0])))
```

```python

"""
Function
--------
switch_guess

The strategy that always switches a guess after the goat door is opened

Parameters
----------
guesses : array
     Array of original guesses, for each simulation
goatdoors : array
     Array of revealed goat doors for each simulation

Returns
-------
The new door after switching. Should be different from both guesses and goatdoors

Examples
--------
>>> print switch_guess(np.array([0, 1, 2]), np.array([1, 2, 1]))
>>> array([2, 0, 0])
"""

def switch_guess(guesses, goatdoors):
    switch = np.random.randint(3, size=guesses.size)
    
    while True:
        bad = (switch == guesses) | (switch == goatdoors)
        if not bad.any():
            return switch
        switch[bad] = np.random.randint(3, size=bad.sum())
    
```

```python
"""
Function
--------
win_percentage

Calculate the percent of times that a simulation of guesses is correct

Parameters
-----------
guesses : array
    Guesses for each simulation
prizedoors : array
    Location of prize for each simulation

Returns
--------
percentage : number between 0 and 100
    The win percentage

Examples
---------
>>> print win_percentage(np.array([0, 1, 2]), np.array([0, 0, 0]))
33.333
"""
#your code here
def win_percentage(guesses, prizedoors):
    return 100 * (guesses == prizedoors).mean()
```

```python
sim = 10000

print("Win percentage when keeping same guess:")
print(win_percentage(simulate_guess(sim), simulate_prizedoor(sim)))

print("Win percentage when changing guess:")
prize = simulate_prizedoor(sim)
guess = simulate_guess(sim)
goat = goat_door(prize, guess)
switch = switch_guess(guess, goat)
print(win_percentage(switch, prize))
```
